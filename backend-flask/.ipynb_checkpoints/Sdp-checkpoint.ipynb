```python
##axis=1 refers to the columns axis. When you set axis=1, it means you are instructing pandas to drop or perform the operation along the columns.
##axis=0, it would mean you want to drop a row along the rows axis.
```

```python
a = df.drop(columns = 'Outcome' , axis = 1)
b = df['Outcome']
```

```python
a.shape
```

```python
b.shape
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(a , b , test_size = 0.2 , random_state = 20)
```

```python
print('Training Data : ' , X_train.shape , Y_train.shape)
print('Testing Data : ' , X_test.shape , Y_test.shape)
```

```python
## training our model logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(
    ##penalty='l2',       # L2 regularization (default)
    ##3C=1.0,              # Regularization strength (default)
    ##slver='lbfgs',     # Optimization algorithm (default)
    #max_iter=100,       # Maximum number of iterations (default)
    ##multi_class='auto', # Auto choose one-vs-rest or multinomial (default)
    ##class_weight=None,  # No class weights (default)
    ##random_state=42,    # Random seed for reproducibility
    ##fit_intercept=True, # Calculate the intercept (default)
    ##intercept_scaling=1 # No scaling of the intercept (default))
)
```

```python
model.fit(X_train, Y_train)
```

```python
output_prediction = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, output_prediction)
```

```python
##accuracy_score : from the sklearn.metrics  calculates the accuracy of the model predictions. The accuracy is defined
##Accuracy = No of Correct pred / Total no of Pred
## returns a value between 0 and 1, where 1 indicates perfect accuracy all predictions
 ## are correct, and 0 indicates no accuracy none of the predictions are correct.
```

```python
##This line prints the accuracy of the model on the test set as a percentage.
##The accuracy variable calculated using the accuracy_score function contains a value between 0 and 1,
## where 1 represents 100% accuracy. Multiplying it by 100 converts it to a percentage for easier interpretation.

## The classification_report is useful for understanding the performance of a classification model,
##especially in scenarios where there are multiple classes.

print("accuracy :", accuracy *100)
print("Classification Report:\n", classification_report(Y_test, output_prediction))
```

```python
## decision tree
```
